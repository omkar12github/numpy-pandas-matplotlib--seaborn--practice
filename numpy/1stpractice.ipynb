```python
import time
```

```python
x = range(100000000) # Create a range object 'x' containing numbers from 0 to 999999
y = range(100000000,20000000) # Create a range object 'y' containing numbers from 1000000 to 1199999

start_time = time.time() # Record the current time before starting the operation

c = [(x, y ) for x,y in zip(x,y)] # Create a list of tuples by zipping together the elements from 'x' and 'y'

print(time.time() - start_time) # Calculate and print the time taken to execute the list comprehension operation
```

```python
a = np.arange(100000000) # Create a numpy array 'y' containing numbers from 0 to 999999
b = np.arange(100000000,200000000) # Create a numpy array 'y' containing numbers from 1000000 to 1999999

start_time = time.time() # Record the current time before starting the operation

c = a + b # Perform element-wise addition of numpy arrays 'a' and 'b'

print(time.time() - start_time) # Calculate and print the time taken to execute the numpy array addition operation
```

INDEXING, SLICING AND ITERATION

```python
arr12 = np.arange(24).reshape(6,4) #np.arange(24) creates a 1D numpy array with values from 0 to 23, and .reshape(6,4) reshapes it into a 2D array with 6 rows and 4 columns
arr12
```

```python
arr12[2] # Access the third row of the 2D numpy array arr12
```

```python
arr12[:2] # Access the first two rows of the 2D numpy array arr12
```

```python
arr12[:,2] # Access the third column of the 2D numpy array arr12 because the colon (:) before the comma indicates all rows, and the 2 after the comma specifies the third column (indexing starts from 0).
```

```python
arr12[:,1:3]
```

```python
arr12[2:4, 1:3] # Access a subarray from the 2D numpy array arr12 that includes rows 2 to 3 (index 2 and 3) and columns 1 to 2 (index 1 and 2).
```

```python
arr12[4:,2:]    # Access a subarray from the 2D numpy array arr12 that includes rows 4 to the end and columns 2 to the end.
```

```python
for i in arr12:
    print(i) # Iterate through each row of the 2D numpy array arr12 and print it
```

```python
for i in np.nditer(arr12):
    print(i) # Iterate through each element of the 2D numpy array arr12 and print it
```
