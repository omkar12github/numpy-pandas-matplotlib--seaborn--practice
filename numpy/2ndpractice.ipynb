INDEXING WITH BOOLEAN ARRAYS

```python
import numpy as np
```

```python
arr = np.random.randint(low = 1, high = 100, size = 20).reshape(4,5)
arr
```

```python
arr[0]
```

```python
arr>50
```

```python
arr[arr>50] #indexing using boolean array
```

```python
arr[(arr>50) & (arr%2!=0)] #indexing using boolean array with multiple conditions
```

```python
arr[(arr>50) & (arr%2!=0)] = 0 #assigning new value to the indexed array
arr
```
