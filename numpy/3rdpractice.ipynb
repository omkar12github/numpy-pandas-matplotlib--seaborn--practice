```python
import numpy as np
```

```python
x = np.linspace(-40,40,100) #linspace(start, stop, num) - generates num evenly spaced values from start to stop
```

```python
x.size
```

```python
y= np.sin(x)
```

```python
y
```

```python
import matplotlib.pyplot as plt #importing the pyplot module from the matplotlib library and giving it an alias plt , pyplot is a collection of functions that make matplotlib work like MATLAB and it is also called sublibrary of matplotlib
%matplotlib inline 
```

```python
plt.plot(x,y)
```
