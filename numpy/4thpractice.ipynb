BROADCASTING

```python
import numpy as np
```

```python
#scenario 1
a1 = np.arange(8).reshape(2,4)
a2 = np.arange(8,16).reshape(2,4) 

print(a1)
print(a2)
```

```python
a1 + a2
```

```python
#scenario 2
a3= np.arange(9).reshape(3,3)
a4 = np.arange(3).reshape(1,3)
print(a3,a4)
```

```python
a3 + a4
```
